**2. Code Program perhitungan manual (menggunakan dimensi 10x10) dari operasi aritmatika (penjumlahan, pengurangan, perkalian, pembagian) dan geometri (translasi (m,n) = (30, 25), rotasi (90 searah jarum jam), flipping vertikal, flipping horisontal).**

```python
import numpy as np
# Manual calculations for a 10x10 image (example)

# Assuming the 10x10 grayscale image is represented by the following matrix:
img_resized3 = np.array([
    [10, 12, 15, 18, 20, 22, 25, 28, 30, 32],
    [11, 13, 16, 19, 21, 23, 26, 29, 31, 33],
    [12, 14, 17, 20, 22, 24, 27, 30, 32, 34],
    [13, 15, 18, 21, 23, 25, 28, 31, 33, 35],
    [14, 16, 19, 22, 24, 26, 29, 32, 34, 36],
    [15, 17, 20, 23, 25, 27, 30, 33, 35, 37],
    [16, 18, 21, 24, 26, 28, 31, 34, 36, 38],
    [17, 19, 22, 25, 27, 29, 32, 35, 37, 39],
    [18, 20, 23, 26, 28, 30, 33, 36, 38, 40],
    [19, 21, 24, 27, 29, 31, 34, 37, 39, 41]
])


# 1. Arithmetic Operations:

# a) Addition (add 10 to each pixel)
img_add = img_resized3 + 10
print("Addition:\n", img_add)

# b) Subtraction (subtract 5 from each pixel)
img_sub = img_resized3 - 5
print("Subtraction:\n", img_sub)

# c) Multiplication (multiply each pixel by 2)
img_mul = img_resized3 * 2
print("Multiplication:\n", img_mul)

# d) Division (divide each pixel by 2)
img_div = img_resized3 / 2
print("Division:\n", img_div)


# 2. Geometric Operations:

# a) Translation (m,n) = (30, 25)
#  (Not applicable for this example as it's just a small matrix,
#   but the concept is to shift the image by the given values)

# b) Rotation (90 degrees clockwise)
img_rot = np.rot90(img_resized3, k=1)
print("Rotation:\n", img_rot)

# c) Vertical Flipping
img_flip_vert = np.flipud(img_resized3)
print("Vertical Flipping:\n", img_flip_vert)

# d) Horizontal Flipping
img_flip_horiz = np.fliplr(img_resized3)
print("Horizontal Flipping:\n", img_flip_horiz)

```

**3. Code Program perhitungan manual (menggunakan dimensi 10x10) histogram dan ekualisasi histogram.**

```python
import matplotlib.pyplot as plt
import numpy as np
# 3. Histogram and Histogram Equalization:

# a) Histogram Calculation (manual for 10x10)
# Create a histogram (frequency of each gray level)
histogram = [0] * 256  # Assuming 8-bit grayscale image
for row in img_resized3:
    for pixel in row:
        histogram[pixel] += 1

# Display the histogram
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.bar(range(256), histogram)
plt.title('Histogram')
plt.xlabel('Gray Level')
plt.ylabel('Frequency')

# b) Histogram Equalization (manual for 10x10)
# Calculate the cumulative distribution function (CDF)
total_pixels = img_resized3.size
cdf = [0] * 256
for i in range(256):
    cdf[i] = sum(histogram[:i+1]) / total_pixels

# Apply the CDF to the image to equalize it
equalized_image = np.zeros_like(img_resized3)
for i in range(img_resized3.shape[0]):
    for j in range(img_resized3.shape[1]):
        equalized_image[i, j] = round(cdf[img_resized3[i, j]] * 255)

# Calculate the histogram of the equalized image
equalized_histogram = [0] * 256
for row in equalized_image:
    for pixel in row:
        equalized_histogram[int(pixel)] += 1

# Display the histogram of the equalized image
plt.subplot(1, 2, 2)
plt.bar(range(256), equalized_histogram)
plt.title('Equalized Histogram')
plt.xlabel('Gray Level')
plt.ylabel('Frequency')

plt.show()
```
